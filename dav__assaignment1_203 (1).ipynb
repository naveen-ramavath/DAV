```python
import pandas as pd
df = pd.read_csv(r"/Users/bomminapradeep/Downloads/student_data.csv")
print("DataSet info: ")
df.info()
```

UNIT I: NumPy Operations
1. Fixed Type Arrays & Creating Arrays

```python
import numpy as np

# Creating a NumPy array from student ages
ages = np.array(df["age"])
print(ages)
```

2. Array Indexing & Slicing

```python
# Indexing - Get the first student's age
print(ages[0])

# Slicing - Get ages of the first 10 students
print(ages[:10])
```

3. Reshaping Arrays

```python
# Reshaping into 2D array (if total elements allow it)
reshaped_ages = ages.reshape(-1, 1)
print(reshaped_ages)
```

4. Array Concatenation & Splitting

```python
# Concatenation of age and absences
absences = np.array(df["absences"])
combined = np.concatenate((ages.reshape(-1,1), absences.reshape(-1,1)), axis=1)
print(combined)

# Splitting the combined array
split1, split2 = np.hsplit(combined, 2)
print(split1, split2)
```

5. Universal Functions

```python
# Applying square root to absences
sqrt_absences = np.sqrt(absences)
print(sqrt_absences)
```

6. Aggregations

```python
import numpy as np
print("Mean Age:", np.mean(ages))
print("Max Absences:", np.max(absences))
print("Sum of Absences:", np.sum(absences))
```

7. Broadcasting Rules

```python
# Subtract mean from all ages
mean_age = np.mean(ages)
normalized_ages = ages - mean_age
print(normalized_ages)
```

8. Comparisons, Boolean Arrays, and Masks

```python
# Finding students older than 16
older_students = ages > 16
print(older_students)
```

9. Fancy Indexing

```python
# Getting ages of students with more than 5 absences
high_absence_students = ages[absences > 5]
print(high_absence_students)
```

10. Sorting

```python
sorted_ages = np.sort(ages)
print(sorted_ages)

sorted_indices = np.argsort(ages)
print(sorted_indices)
```

UNIT II: Pandas Operations
1. Series & DataFrame

```python
# Creating Series
age_series = pd.Series(df["age"])
print(age_series)

# Creating a DataFrame with selected columns
df_selected = df[["age", "absences", "G3"]]
print(df_selected.head())
```

2. Indexing & Selecting

```python
# Selecting specific column
print(df["age"])

# Selecting specific row
print(df.iloc[0])

# Selecting multiple columns
print(df.loc[:, ["age", "absences"]])
```

3. Universal Functions

```python
# Applying a function to a column
df["age_squared"] = df["age"].apply(lambda x: x**2)
print(df.head())
```

4. Handling Missing Data

```python
import numpy as np

# Check for missing values
print("Missing values before filling:\n", df.isnull().sum())

# Fill missing values only in numeric columns
df.fillna(df.select_dtypes(include=np.number).mean(), inplace=True)

# Verify that missing values are filled
print("Missing values after filling:\n", df.isnull().sum())
```

5. Hierarchical Indexing

```python
df.set_index(["school", "sex"], inplace=True)
print(df.head())
```

UNIT III: Combining Datasets
1. Concatenation & Appending

```python
df_copy = df.copy()
df_concat = pd.concat([df, df_copy], axis=0)
print(df_concat.shape)
```

2. Merging & Joins

```python
import pandas as pd
import numpy as np

# Load dataset from the uploaded file
file_path = "/Users/bomminapradeep/Downloads/student_data (2).csv"  # Change this if needed
df = pd.read_csv(file_path)

# Ensure column names are clean
df.columns = df.columns.str.strip()

# Check if 'school' column exists
if "school" in df.columns:
    # Fill missing values in 'school'
    df["school"] = df["school"].fillna("Unknown")

    # Create a second DataFrame for merging
    unique_schools = df["school"].dropna().unique()
    df_extra = pd.DataFrame({
        "school": unique_schools,
        "extra_activity": np.random.choice(["Sports", "Music", "Art"], size=len(unique_schools))
    })

    # Merge the DataFrames
    df_merged = df.merge(df_extra, on="school", how="left")

    # Display results
    print(df_merged.head())
else:
    print("Error: 'school' column not found in dataset. Available columns:", df.columns)
```

3. Aggregation & Grouping

```python
# Group by school and calculate average grade
grouped = df.groupby("school")["G3"].mean()
print(grouped)
```

4. Pivot Tables

```python
pivot = df.pivot_table(values="G3", index="school", columns="sex", aggfunc="mean")
print(pivot)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set_style("whitegrid")

# ================================
# 📌 Load Dataset
# ================================
df = pd.read_csv("student_data.csv")  # Change to the correct filename if needed

# ================================
# 📌 1️⃣ Scatter Plot (G1 vs G3)
# ================================
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['G1'], y=df['G3'])
plt.title("Scatter Plot: G1 vs G3 (First Grade vs Final Grade)")
plt.xlabel("G1 (First Grade)")
plt.ylabel("G3 (Final Grade)")
plt.show()

# ================================
# 📌 2️⃣ Bar Plot (Study Time vs Final Grade)
# ================================
plt.figure(figsize=(6,4))
sns.barplot(x=df['studytime'], y=df['G3'])
plt.title("Bar Plot: Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()

# ================================
# 📌 3️⃣ Heatmap (Correlation Matrix)
# ================================
plt.figure(figsize=(10,6))
sns.heatmap(df[['age', 'studytime', 'failures', 'G1', 'G2', 'G3']].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation Between Numeric Columns")
plt.show()

# ================================
# 📌 4️⃣ Pie Chart (Gender Distribution)
# ================================
plt.figure(figsize=(6,6))
df['sex'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.title("Pie Chart: Gender Distribution")
plt.show()

# ================================
# 📌 5️⃣ Histogram (Distribution of Final Grades)
# ================================
plt.figure(figsize=(6,4))
plt.hist(df['G3'], bins=10, color='green', alpha=0.7)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Frequency")
plt.title("Histogram: Distribution of Final Grades")
plt.show()
```
